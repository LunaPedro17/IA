```python
#tensorflow/pytorch
import pandas as pd

datos= pd.read_csv("celsius.csv")
datos.head()
```

```python
datos["celsius"].values
```

```python
x=datos["celsius"].values
y=datos["farenheit"].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential(
    [
        tf.keras.layers.Input(shape=(1,)),#la entrada, los celcius: 1 solo datos
        tf.keras.layers.Dense(units=1),#la salida, tambien un datos, grados fahrenheit
    ]
)
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),#parametro es la tasa de aprendizaje
    loss='mean_squared_error'
    )
```

```python
entrenamiento= modelo.fit(x,y,epochs=1000)#epocas cuantas vueltas da al set de datos
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(entrenamiento.history["loss"])
```

```python
import numpy as np

valor = 1230
# Convert the input to a NumPy array
prediccion = modelo.predict(np.array([valor]))
print(f"Predicción: {valor} celsius son {prediccion[0]} fahrenheit")
```
